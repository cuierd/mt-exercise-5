```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
d ={
    "beam_size" : [1,2,3,4,5,6,8,10,12,15,20],
    "bleu":[22.8,23.9,23.9,24.1,24.3,24.4,24.4,24.4,24.4,24.4,24.3]
    }
```

```python
df = pd.DataFrame(data=d)

sns.set_theme(style="darkgrid")
lineplot = sns.lineplot(data=df,x="beam_size",y="bleu")
fig = lineplot.get_figure()
fig.savefig("bleu_beam.png",dpi=300)
```
